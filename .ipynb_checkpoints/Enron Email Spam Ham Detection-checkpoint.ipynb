# Enron Spam Ham Email Detection Algotrithm

## Load the Data

Data comes in 6 zip files. Witin each are 2 folders - one with spam emails and one with ham emails. First task is to import these emails into a dataframe with the spam/ham labels.

```python
import pandas as pd
import sklearn as sk
import numpy as np
import os
```

First create a 2 lists of all the relevent spam and ham text files we will be importing

```python
def emaillist(dir):
    files = []
    for file in os.listdir(dir):
        a = dir + file
        files.append(a)
    return files
```

```python
hamfiles = emaillist("./enron1/ham/") + emaillist("./enron2/ham/") + emaillist("./enron3/ham/") + emaillist("./enron4/ham/") + emaillist("./enron5/ham/") + emaillist("./enron6/ham/")
spamfiles = emaillist("./enron1/spam/") + emaillist("./enron2/spam/") + emaillist("./enron3/spam/") + emaillist("./enron4/spam/") + emaillist("./enron5/spam/") + emaillist("./enron6/spam/")
```

```python
hamfiles
```

Import these text documents into a list of every email.

```python
def importfiles(list):
    emaillist = []
    for file in list:
        with open(file, encoding="latin-1") as f:
            a = f.read()
            emaillist.append(a)
    return emaillist
```

```python
hamemails = importfiles(hamfiles)
spamemails = importfiles(spamfiles)
```

```python
print (hamemails[3])
```

Now to put these in a dataframe and add labels showing whether spam or ham.

```python
spamdf = pd.DataFrame({'email':spamemails,'class':"spam"})
hamdf = pd.DataFrame({'email':hamemails,'class':"ham"})
emaildf = pd.concat([spamdf, hamdf], ignore_index=True)
```
